```python
# !python3
# -*- coding:utf-8 -*-
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn.datasets
import sklearn.model_selection
# %matplotlib inline



def loss_fuc(x,y,w):
    #print (np.shape(x),np.shape(y),np.shape(w))
    loss=(np.dot(x,w)-y)
    loss_len=loss.shape[0]
    loss=np.sum(np.dot(loss.T,loss))
    loss=(loss/(2*loss_len))
    return loss

def optimizer(x,y,xvali,yvali,w,lr,n_estimator):
    xlen=len(x)
    loss_train=[]
    loss_vali=[]
    for i in range(n_estimator):
        tloss=loss_fuc(x, y, w)
        print ("[The %d iteration in train]: %f" % (i,tloss))
        vloss=loss_fuc(xvali,yvali,w)
        print("[The %d iteration in validation]: %f" % (i,vloss))
        loss_train.append(tloss)
        loss_vali.append(vloss)
        pred=np.dot(x,w)
        #print ("pred:",np.shape(pred))
        # delta=pred - y
        #print ("delta:",np.shape(delta))
        w_gradient = np.dot(x.T, pred - y)/xlen
        #print (np.shape(w_gradient))
        w=w-lr*w_gradient
    return w,loss_train,loss_vali




def Linear_regression(x,y,xvali,yvali):
    n_sample,n_feature=np.shape(x)
    n_vali_sample,n_vali_feature=np.shape(xvali)
    lr=0.001
    w0=np.zeros(shape=(n_feature+1,1))
    x=np.append(np.ones(shape=(n_sample,1)),x,1)
    xvali=np.append(np.ones(shape=(n_vali_sample,1)),xvali,1)
    n_estimator=10000

    w,loss_train,loss_vali=optimizer(x,y,xvali,yvali,w0,lr,n_estimator)
    return loss_train,loss_vali

```

```python
X, y = sklearn.datasets.load_svmlight_file('../data/housing_scale')
X=X.todense()
l=len(y)
y=y.reshape(l,1)
#print ('X',X)
#print ('y',y)
X_train, X_vali, y_train, y_vali = sklearn.model_selection.train_test_split(X, y, random_state=1000)
#print ('x_vali:',X_vali)
loss_train,loss_vali=Linear_regression(X_train,y_train,X_vali,y_vali)
```

```python
plt.plot(loss_train,label='Ltrain')
plt.plot(loss_vali,label='Lvalidation')
plt.title('Ltrain and Lvalidation loss')
plt.xlabel('iteration number')
plt.ylabel('datasets loss')
plt.legend()
plt.show()
```
